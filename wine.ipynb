# Histogram with kde plot

```python
for i, col in enumerate(wine.columns):
    print(i,col)
    plt.figure(i)
    sns.distplot(wine[col],kde=True,hist_kws={'edgecolor':'black'},bins=20)
```

```python
for x,col in enumerate(wine.columns):
    plt.figure(x)
    sns.boxplot(wine[col])
```

# Multivariate analysis

```python
#corelation matrix
cor=wine.corr()
cor
plt.subplots(figsize=(20,15))
sns.heatmap(cor,annot=True,fmt=".2g",vmin=-1,vmax=1,center=0,cmap="inferno",annot_kws={"size":15})
plt.xticks(size=20)
plt.yticks(size=20)
```

By looking in the plot we can see that:
1-The variables with nearly negligible relation or no relation with quality are-
pH,free sulphur dioxide and citric acid.
2-The variable with negative relationship are-
fixed acidity,volatile acidity,residual sugar, chlorides,total sulphur dioxide,density
3-The variable with positive relationship are-
sulphates and alcohol

So by looking at the data we can say that the quality of wine will get better when-

The amount of negative related variable will be low and the amount of positive realted variable will be high
